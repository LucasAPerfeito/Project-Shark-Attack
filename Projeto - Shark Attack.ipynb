```python
df3_fem = df3[df3['sex']=='F'].groupby('country')[['year']].count().sort_values(by = 'year', ascending = False).head(10).fillna(0)
```

```python
df3_masc = df3[df3['sex']=='M'].groupby('country')[['year']].count().sort_values(by = 'year', ascending = False).head(10).fillna(0)
```

```python
df3_total = pd.merge(df3_fem, df3_masc, how = 'outer', right_index = True, left_index = True, suffixes = ('_f','_m')).fillna(0)
```

```python
df3_total['proporcao_f'] = (df3_total['year_f']/(df3_total['year_m'] + df3_total['year_f']))*100
df3_total.sort_values('proporcao_f', ascending = False)
```
